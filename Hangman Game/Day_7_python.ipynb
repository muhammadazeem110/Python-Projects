- ##### First step is :
    1) Make a list of some words form English dictionary and make a list of all
    2) Choose a random word from the list
    3) For each letter in choose_word, add a '_' to guess_word which will store the all true inputs from users

```python
#import lilbrary

import random
import hangman_words
from hangman_words import words_list

choose_word = random.choice(words_list)
guess_word = []
for i in range(0, len(choose_word)):
    guess_word += '_'
```

- ##### Second step is :
    4) Run the loop until it all the words from choose_word is not stored in guess_word list
    5) Get a letter from user
    6) Now compare the input letter with the each letter in choose_word
    7) If letters match then replace with them else hang a part of hangman

```python
lives = 6
end_of_game = True
while end_of_game:
    guess_letter = input("Guess the letter : ").lower()
    if guess_letter in guess_word:
        print(f"You already guess {guess_letter} in the word.")

    for position in range(0, len(choose_word)):
        if guess_letter == choose_word[position]:
            guess_word[position] = guess_letter
    
    if guess_letter not in choose_word:
        print(f"You guess {guess_letter} is not in the word. You lose a life.")
        lives -= 1

    if '_' not in guess_word:
        print("Hooray! You wonðŸŽ‰")
        end_of_game = False
    
    print(f"{' '.join(guess_word)}")
    
    if lives == 0:
        print("You lose the Game.")
        end_of_game = False
        
    from hangman_art import stages
    print(stages[lives])
```
